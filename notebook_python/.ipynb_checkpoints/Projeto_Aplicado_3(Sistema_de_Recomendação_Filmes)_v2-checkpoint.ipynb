```python
#Importando os pacotes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchCV
from boruta import BorutaPy
import xgboost as xgb
from imblearn.over_sampling import SMOTE

%matplotlib inline
sns.set_style('darkgrid')
```

### 5.1 Tabela com os metadados

```python

# extrair nome das colunas
nomes_colunas = df.columns.tolist()

# extrair tipos de dados
tipos_dados = df.dtypes.tolist()

# extrair número de valores nulos
num_nulos = df.isnull().sum().tolist()

# criar lista com exemplos de valores e origem dos dados
exemplos = []
for coluna in nomes_colunas:
    exemplo = df[coluna][0]
    exemplos.append(exemplo)
origem = ['ethereum_transaction_dataset'] * len(nomes_colunas)

# criar DataFrame com as informações extraídas
metadata = pd.DataFrame({
    'Variável': nomes_colunas,
    'Tipo': tipos_dados,
    'Exemplo': exemplos,
    '% Nulos': [round((nulos / df.shape[0]) * 100, 2) for nulos in num_nulos],
    'Origem': origem
})

# exibir tabela de metadados
metadata
```

### 5.6 Preparação dos dados



##### Aplicação dos métodos de seleção de variáveis;  - Boruta

```python
#Separamos em X e y para usar o Boruta
X = df.drop("FLAG", axis=1)
y = df["FLAG"]
```

```python
# Divindo 30% para teste e 70% para treino
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30,
                                                          random_state = 43)
```

```python
y = y_train.ravel()
```

```python
# Definindo os valores a serem testados para cada parâmetro
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20],
    'max_features': ['sqrt', 'sqrt', 'log2']
}

# Instanciando o modelo RandomForestClassifier
rfc = RandomForestClassifier()

# Instanciando o GridSearchCV com o modelo e os valores a serem testados
grid_search = GridSearchCV(estimator=rfc, param_grid=param_grid, cv=5)

# Treinando o modelo com a busca em grade
grid_search.fit(X_train, y_train)

# Exibindo os melhores valores para os parâmetros
print("Melhores valores para n_estimators, max_depth e max_features:", grid_search.best_params_)
```

```python
rf = RandomForestClassifier(random_state = 43, max_depth = 20, max_features = 'sqrt', n_estimators = 200, bootstrap = True)
```

```python
feat_selector = BorutaPy(rf, n_estimators = "auto", random_state = 43)
```

```python
np.int = np.int32
np.float = np.float64
np.bool = np.bool_
result_boruta = feat_selector.fit(X_train.values, y_train)
```

```python
# Verificar as features selecionadas
accept = X_train.columns[feat_selector.support_].to_list()
irresolution = X_train.columns[feat_selector.support_weak_].to_list()

print("Selecionadas:")
print("************************")
print(list(accept))
print(list(irresolution))
print("************************")
```

```python
# .support_
result_boruta.support_
```

```python
## removendo as colunas rejeitadas pelo boruta
result_boruta.support_ = np.insert(result_boruta.support_, 0, True) ## adcionando um true na coluna FLAG

df.drop(df.columns[result_boruta.support_ == False], axis=1, inplace=True)
```

### 5.7 Particionamento dos dados em treino, validação e teste

```python
#Separamos em X e y
X = df.drop("FLAG", axis=1)
y = df["FLAG"]
```

```python
X
```

```python
y
```

```python
# Divindo 30% para teste e 70% para treino
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30,
                                                          random_state = 43)
```

### 5.8 Lidando com o desbalaceamento


```python
oversample = SMOTE()
print(f'Antes do balanceamento: {X_train.shape, y_train.shape}')

x_tr_resample, y_tr_resample = oversample.fit_resample(X_train, y_train)
print(f'Depois do balanceamento: {x_tr_resample.shape, y_tr_resample.shape}')
```

```python
# Target distribução antes balanceamento
non_fraud = 0
fraud = 0

for i in y_train:
    if i == 0:
        non_fraud +=1
    else:
        fraud +=1

# Target distribução depois  balanceamento
no = 0
yes = 1

for j in y_tr_resample:
    if j == 0:
        no +=1
    else:
        yes +=1


print(f'Antes Balanceamento \n \tNon-frauds: {non_fraud} \n \tFauds: {fraud}')
print(f'Depois Balanceamento \n \tNon-frauds: {no} \n \tFauds: {yes}')
```

### 5.9 Seleção dos modelos a serem aplicados e hiperparâmetros a serem testados;
### Modelo 1 - Random Forest


```python
# Definindo os valores a serem testados para cada parâmetro
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20],
    'max_features': ['sqrt']
}

# Instanciando o modelo RandomForestClassifier
rfc = RandomForestClassifier()

# Instanciando o GridSearchCV com o modelo e os valores a serem testados
grid_search = GridSearchCV(estimator=rfc, param_grid=param_grid, cv=5)

# Treinando o modelo com a busca em grade
grid_search.fit(x_tr_resample, y_tr_resample)

# Exibindo os melhores valores para os parâmetros
print("Melhores valores para n_estimators, max_depth e max_features:", grid_search.best_params_)
```

```python
rf = RandomForestClassifier(random_state = 43, n_estimators = 100, max_depth = 20,
                            max_features = 'sqrt').fit(x_tr_resample, y_tr_resample)
```

```python
y_train_pred = rf.predict(X_train)
y_test_pred = rf.predict(X_test)
```

```python
#Calculo da acurácia do treino
train_accuracy = accuracy_score(y_train, y_train_pred)

#Calculo da acurácia do teste
test_accuracy = accuracy_score(y_test, y_test_pred)

print('% Acertos treino:', train_accuracy)
print('% Acertos teste:', test_accuracy)
```

```python
train_AUC = roc_auc_score(y_train, y_train_pred)
test_AUC  = roc_auc_score(y_test, y_test_pred)

print('% AUC:', train_AUC, test_AUC)
```

```python
y_train_pred = rf.predict(X_train) #Pred Train
y_test_pred = rf.predict(X_test) #Pred Test

print(classification_report(y_test, y_test_pred))
print(confusion_matrix(y_test, y_test_pred))
```

```python
confusion_matrix0 = metrics.confusion_matrix(y_train, y_train_pred)
confusion_matrix1 = metrics.confusion_matrix(y_test, y_test_pred)

print(confusion_matrix0, "\n","\n",confusion_matrix1)
```

### Modelo 2 - XGBoosting

```python
X = df.drop("FLAG", axis=1)
y = df["FLAG"]

# Divindo 30% para teste e 70% para treino
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30,
                                                          random_state = 43)

oversample = SMOTE()


x_tr_resample, y_tr_resample = oversample.fit_resample(X_train, y_train)



xgb_c = xgb.XGBClassifier(random_state=43)
```

```python
# Definindo os valores a serem testados para cada parâmetro
param_grid = {
    'learning_rate': [0.01, 0.1, 0.5],
    'n_estimators': [100, 200],
    'max_depth': [3, 4, 5],
    'subsample': [0.3, 0.5, 0.9],
    'colsample_bytree': [0.3, 0.5, 0.7]
}


# Instanciando o GridSearchCV com o modelo e os valores a serem testados
grid_search = GridSearchCV(estimator=xgb_c, param_grid=param_grid, cv=5, n_jobs=-1)


# Treinando o modelo com a busca em grade
grid_search.fit(x_tr_resample, y_tr_resample)

# Exibindo os melhores valores para os parâmetros
print("Melhores valores para learning_rate, n_estimators, max_depth, subsample e colsample_bytree:", grid_search.best_params_)
```

```python
clf_xgb = xgb.XGBClassifier(random_state = 43, colsample_bytree = 0.7, learning_rate = 0.1, max_depth = 5,
                            n_estimators = 200, subsample = 0.5)
```

```python
clf_xgb.fit(x_tr_resample, y_tr_resample)
```

```python
y_train_pred = clf_xgb.predict(X_train) #Pred Train
y_test_pred = clf_xgb.predict(X_test) #Pred Test
```

```python
#Calculo da acurácia do treino
train_accuracy = accuracy_score(y_train, y_train_pred)

#Calculo da acurácia do teste
test_accuracy = accuracy_score(y_test, y_test_pred)

print('% Acertos treino:', train_accuracy)
print('% Acertos teste:', test_accuracy)
```

```python
print(classification_report(y_test, y_test_pred))
print(confusion_matrix(y_test, y_test_pred))
```

```python
train_AUC = roc_auc_score(y_train, y_train_pred)
test_AUC  = roc_auc_score(y_test, y_test_pred)

print('% AUC:', train_AUC, test_AUC)
```

```python
confusion_matrix0 = metrics.confusion_matrix(y_train, y_train_pred)
confusion_matrix1 = metrics.confusion_matrix(y_test, y_test_pred)

print(confusion_matrix0, "\n","\n",confusion_matrix1)
```

## Modelo Escolhido XGBoosting
#### Esse modelo trouxe uma melhor performance desde acurácia no teste, com um desvio menor que a Random Forest(treino x teste, e o resultado se prova ao analisar a curva ROC e matriz de confusão.

```python
modelo = clf_xgb
importances  = modelo.feature_importances_
total_importances = sum(importances)
importances_pct = [(i / total_importances) * 100 for i in importances]

# Criando um novo DataFrame com as feature importances em porcentagem e os nomes das colunas
df_importances = pd.DataFrame({'coluna': X.columns, 'importancia_pct': importances_pct})

# Ordenando as feature importances em ordem decrescente
df_importances = df_importances.sort_values('importancia_pct', ascending=False)

print(df_importances)
```
